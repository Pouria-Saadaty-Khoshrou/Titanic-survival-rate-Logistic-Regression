```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_data = pd.read_csv('titanic_train.csv')
```

```python
sns.boxplot(x='Pclass',y='Age',data=train_data)
```

```python
def age_changer(data):
    
    age = data[0]
    pclass = data[1]
    
    if pd.isnull(age):
        if pclass == 1:
            age = 37
        elif pclass == 2:
            age = 29
        else:
            age = 23
    return age
```

```python
#train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
train_data['Age'] = train_data[['Age','Pclass']].apply(age_changer, axis=1)
```

```python
train_data.head()
```

```python
train_data = pd.concat([train_data,Sex,Embarked],axis=1)
```
